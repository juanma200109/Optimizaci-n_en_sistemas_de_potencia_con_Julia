### 3. Personalización de Gráficos

Personalizar los gráficos mejora la claridad y el impacto visual.

#### **Títulos y Etiquetas**

```julia
using Plots
x = 1:5
y = [100, 150, 200, 180, 160]
plot(x, y, 
     title="Generación por Nodo", 
     xlabel="Nodo", 
     ylabel="Potencia (MW)", 
     label="PGEN")
```

**Leyendas y Colores**

```julia
plot(x, y, 
     label="Potencia Generada", 
     color=:purple, 
     linewidth=2)
```

#### Múltiples Series

Puedes graficar varias series en el mismo gráfico:

```julia
y2 = [95, 148, 192, 178, 160]
plot(x, y, label="PGEN", color=:blue)
plot!(x, y2, label="PLOAD", color=:red) # plot! añade a la gráfica existente
```

Nota: plot! agrega una nueva serie al gráfico existente.
