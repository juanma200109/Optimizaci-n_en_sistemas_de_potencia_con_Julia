### 4. Gráficos en Sistemas de Potencia

En el contexto de sistemas de potencia, los gráficos son esenciales para:





* Visualizar la generación (PGEN) y carga (PLOAD) por nodo.
* Mostrar los flujos de potencia en líneas (pL).
* Comparar costos operativos.

Ejemplo Práctico:Supongamos que tienes un `DataFrame` con datos de nodos:

```julia
using Plots
using DataFrames

nodes = DataFrame(
    Nodo = [1, 2, 3, 4],
    PGEN = [100.0, 150.0, 200.0, 50.0],
    PLOAD = [50.0, 60.0, 70.0, 40.0]
)

# Gráfico de barras para PGEN y PLOAD
bar(x = nodes.Nodo, [nodes.PGEN nodes.PLOAD], 
    label=["PGEN" "PLOAD"], 
    title="Generación vs Carga por Nodo", 
    xlabel="Nodo", 
    ylabel="Potencia (MW)")
```
